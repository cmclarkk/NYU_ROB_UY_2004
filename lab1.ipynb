# Simulation

```python
import numpy as np
from rob2004.interface import RobotInterfaceSim
```

```python
robot = RobotInterfaceSim()
```

```python
robot.getJointState()
```

```python
tau = np.zeros(12)
tau[0] = 2.0
robot.setJointCommand(tau)
```

```python
robot.robot.show(np.zeros(12).tolist())
```

# Real Robot

```python
from rob2004.interface import RobotInterfaceReal
robot_real = RobotInterfaceReal("10.20.60.60")
```

```python
robot_real.getJointState()
```

```python
import time
start_time = time.time()
while time.time()-start_time < 10:
    tau = np.zeros(12)
    kd = np.ones(12)*0.1
    robot_real.setJointCommand(tau=tau, kd=kd)
    time.sleep(0.02)
```

```python
import time
start_time = time.time()
while time.time()-start_time < 100:
    state = robot_real.getJointState()
    q = state['q']
    robot.robot.show(q.tolist())
    time.sleep(0.02)
    print(q)
```
